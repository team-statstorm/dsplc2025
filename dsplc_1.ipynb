# **Preprocessing**



```python
#check sum of missing values
df.isnull().sum()
```

Idenifying the sum of missing values

```python
#display where customer id is null
df[df['Customer_ID'].isnull()]
```

Displaying the rows that have null values for customer ID

```python
#drop these missing value on customer ID
df.dropna(subset=['Customer_ID'], inplace=True)
```

Dropping the null values of Customer ID becuase null values of customer id csnnot be assumed using mean or other methods

```python
#display where outlet city is null
df[df['outlet_city'].isnull()]
```

Displaying rows where outlet city is null

```python
#take mode for the outlet_city column and filled missing value with the mode
df['outlet_city'].fillna(df['outlet_city'].mode()[0], inplace=True)
```

Here we have filled the missing values of outlet city using its mode because its a real time dataset.

```python
#display where cluster category is null
df[df['cluster_catgeory'].isnull()]
```

Identifying the rows where cluster category is null

```python
#drop this record
df.dropna(subset=['cluster_catgeory'], inplace=True)
```

Dropping records where cluster category column is null because if the we cant assume cluster category values as its our target variable

```python
#fill missing values in dry sales with median
df['dry_sales'].fillna(df['dry_sales'].median(), inplace=True)
```

here we filled the missing values of dry sales with the median

```python
#fill missing values in fresh sales with median
df['fresh_sales'].fillna(df['fresh_sales'].median(), inplace=True)
```

here we filled the missing values of fresh sales with the median

```python
#replace missing values in luxury sales with '0'
df['luxury_sales'].fillna(0, inplace=True)
```

# Feature Transformation

```python
# Create a dictionary mapping unique cities to their alphabetical order
city_mapping = {city: i+1 for i, city in enumerate(sorted(df['outlet_city'].unique()))}

# Apply the mapping to create a new encoded column
df_scaled['Outlet_No.'] = df['outlet_city'].map(city_mapping)
```

```python
# Rearreange the dataset in ascending order based on Outlet_No.
df_scaled = df_scaled.sort_values('Outlet_No.')
df_scaled.head()
```

Distribution of Customer Transactions by Outlet No.

```python
outlet_transaction_counts = df_scaled['Outlet_No.'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(outlet_transaction_counts.index, outlet_transaction_counts.values)
plt.xlabel("Outlet No.")
plt.ylabel("Number of Transactions")
plt.title("Distribution of Customer Transactions by Outlet No.")
plt.xticks(outlet_transaction_counts.index)
plt.show()
```

Total Luxury Sales per Outlet

```python
outlet_luxury_sales = df_scaled.groupby('Outlet_No.')['luxury_sales'].sum()
plt.figure(figsize=(10, 6))
plt.bar(outlet_luxury_sales.index, outlet_luxury_sales.values)
plt.xlabel("Outlet No.")
plt.ylabel("Total Luxury Sales")
plt.title("Total Luxury Sales per Outlet")
plt.xticks(outlet_luxury_sales.index)
plt.show()
```
