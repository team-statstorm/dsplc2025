```python
numerical_cols = ['luxury_sales', 'fresh_sales', 'dry_sales']

def remove_outliers_iqr(df, column):
    """
    Identifies outliers in a column using the IQR method.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column (str): The column to check for outliers.

    Returns:
        pd.DataFrame: A new DataFrame with the outliers removed.
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return lower_bound, upper_bound

# Example usage with IQR:
filter_condition = pd.Series([True] * len(df))

for col in numerical_cols:
    lower_bound, upper_bound = remove_outliers_iqr(df, col)
    filter_condition = filter_condition & (df[col] >= lower_bound) & (df[col] <= upper_bound)

df = df[filter_condition]

print(f"Dataframe shape after outlier removal using IQR: {df.shape}")

# check the new dataframe
print(df.describe())
```
